```python
def get_songs_artist(user, playlist_id):
    songs_artist = {}
    results = sp.user_playlist_tracks(user, playlist_id)
    for item in results['items']:
        song_name = item['track']['name']
        artist = item['track']['artists'][0]['name']
        song_id = item['track']['uri'].split(':')[-1]
        audio_features = sp.audio_features(song_id)[0]
        songs_artist[song_id] = {'song_name':song_name, 'artist':artist, 'audio_features':audio_features}
    while results['next']:
        results = sp.next(results)
        for item in results['items']:
            song_name = item['track']['name']
            artist = item['track']['artists'][0]['name']
            song_id = item['track']['uri'].split(':')[-1]
            audio_features = sp.audio_features(song_id)[0]
            songs_artist[song_id] = {'song_name':song_name, 'artist':artist, 'audio_features':audio_features}
    return songs_artist
```

```python

import pandas as pd

def get_songs_artist_f(user, playlist_id):
    songs_artist = []
    results = sp.user_playlist_tracks(user, playlist_id)
    for item in results['items']:
        song_name = item['track']['name']
        artist = item['track']['artists'][0]['name']
        song_id = item['track']['uri'].split(':')[-1]
        audio_features = sp.audio_features(song_id)[0]
        row = {'song_id': song_id, 'song_name':song_name, 'artist':artist}
        row.update(audio_features)
        songs_artist.append(row)
    while results['next']:
        results = sp.next(results)
        for item in results['items']:
            song_name = item['track']['name']
            artist = item['track']['artists'][0]['name']
            song_id = item['track']['uri'].split(':')[-1]
            audio_features = sp.audio_features(song_id)[0]
            row = {'song_id': song_id, 'song_name':song_name, 'artist':artist}
            row.update(audio_features)
            songs_artist.append(row)
    return pd.DataFrame(songs_artist)

```
