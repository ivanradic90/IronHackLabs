```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
categoricals = pd.read_csv('categorical.csv')
numericals = pd.read_csv('numerical.csv')
target = pd.read_csv('target.csv')
```

```python
target.head()
```

```python
target = target.drop('TARGET_D', axis=1)
```

```python
target.isna().sum()
```

```python
encoder = OneHotEncoder(drop='first').fit(categoricals)
encoded_categoricals = encoder.transform(categoricals).toarray()
encoded_categoricals = pd.DataFrame(encoded_categoricals)
```

```python
encoded_categoricals
```

```python
data = pd.concat([numericals, encoded_categoricals, target], axis = 1)
```

```python
y = data['TARGET_B']
X = data.drop(['TARGET_B'], axis = 1)
```

```python
X.shape
```

```python
y.shape
```

```python
data['TARGET_B'].value_counts()
```

```python
smote = SMOTE()
X_sm, y_sm = smote.fit_resample(X, y)
y_sm.value_counts()
```

```python
X_sm.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.25, random_state=0)
```

```python
RFC = RandomForestClassifier(max_depth=2, random_state=0)
RFC.fit(X_train, y_train)
print(RFC.score(X_test, y_test))
```
